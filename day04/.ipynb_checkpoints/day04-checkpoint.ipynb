```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
x = pd.read_csv('cancer_expression.csv')
x.head(5)
```

```python
control_cols = [i for i in x.columns if 'Control' in i]
print('Control Columns: ',control_cols)
Treat_cols = [i for i in x.columns if 'Treatment' in i]
print('Control Columns: ',Treat_cols)
```

```python
plt.bar(control_cols,Treat_cols,color = ['blue','g'])
plt.title("Cancer_Expression")
plt.show()
```

```python
gene = "Gene10"
gene_data = x[x["Gene"] == gene]

gene_data

control_mean = gene_data[control_cols].mean(axis=1).values[0]
treatment_mean = gene_data[Treat_cols].mean(axis=1).values[0]

plt.bar("Control",control_mean,color="blue",label="Control")
plt.bar("Treatment",treatment_mean,color="red",label="Treatment")

bars = plt.bar(
     ["Control","Treatment"],[control_mean,treatment_mean],color=["blue","red"]
)


plt.title(f"Bar Chart: {gene} Average Expression")
plt.ylabel("Expression_Level")
plt.legend(bars,["Control","Treatment"],bbox_to_anchor=(1.05,1),loc='upper left')
```

```python
control_data = x[['Control1','Control2']].values.flatten()
treatment_data = x[['Treatment1','Treatment2']].values.flatten()

plt.figure(figsize=(10,6))
sns.boxplot(data = [control_data,treatment_data],palette="Set2")

plt.xticks([0,1],['Control','Treatment'])
plt.title(f"Box plot for {gene}")
plt.ylabel("Expression Levels")
plt.show()
```

```python
gene = "Gene2"
gene_row = x[x['Gene'] == gene].iloc[0]

data = [gene_row[control_cols].values,gene_row[Treat_cols].values]
sns.violinplot(data=data)

plt.xticks([0,1],['Control',"Treatment"])
plt.title(f'Box plot for {gene} Expression istribution')
plt.ylabel('Expression Levels')
plt.show()
```

```python
heatmap_data = x.set_index('Gene')
heatmap = heatmap_data.head(10)

plt.figure(figsize=(12,8))
sns.heatmap(heatmap,cmap="coolwarm",annot=True)

plt.title("Heatmap of Gene Expression Levels")
plt.xlabel("Samples")
plt.ylabel("Genes")
plt.show()
```
